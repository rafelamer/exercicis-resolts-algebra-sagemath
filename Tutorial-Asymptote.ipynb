```python
%%asy --fmt pdf

settings.render = 10;
settings.prc = false;

import three;

currentprojection=perspective((10,8,4));
unitsize(0.6666cm);

/*
    L'equció del pla pl és z=0 i la del pla pl2 és 13y+16z-8=0
*/
path3 pl = (-7,-11,0)--(7,-11,0)--(7,11,0)--(-7,11,0)--cycle;
path3 pl2 = (-7,-8,7)--(7,-8,7)--(7,8,-6)--(-7,8,-6)--cycle;

/*
    Els punts P0 i P1 són del pla pl i els punts Q0 i P1 són de pl2
*/
triple P0 = (5,-9,0),
       P1 = (5,9,0),
       Q0 = (-4.0,-7,6),
       Q1 = (-4.0,7,-5.1875);
draw(P0--P1,1.2bp+black);
draw(Q0--Q1,1.2bp+black);
       
/*
    La recta intersecció dels dos plans és
             y=0.615385
             z=0
    i els punts R0 i R1 són d'aquesta recta. A més,
    aquesta recta és perpendicular a les rectes POP1
    a Q0Q1
*/
triple R0 = (-8.5,0.615385,0),
       R1 = (8.5,0.615385,0);
draw(R0--R1,1.4bp+red);    
draw(surface(pl2),lightgreen+opacity(0.2),1.2bp+.8green);
draw(surface(pl),lightblue+opacity(0.2),1.2bp+.8blue);
label("$P$",(-7,11,0.45),blue);    

/*
    La intersecció de la recta P0P1 amb el pla pl2 és R0 i el de
    la recta Q0Q1 amb el pla pl és R1
*/
triple R2 = (5,0.615385,0),
       R3 = (-4,0.615385,0);
dot(R2,5bp+red);
dot(R3,5bp+red);
label("$Q$",(7,-8,7.45),green);
/*
    Dos punts de la recta POP1
*/
triple P2 = (5,3,0),
       P3 = (5,5.5,0);
dot(P2,4.25bp+black);
draw(P2--P3,2bp+black,Arrow3(6));
label("$p_0$",P2,Z);
label("$\vec u$",(P2+P3)/2,Z);

/*
    La recta Q0Q1 té equació 
          x=-4
          13y+16z-8=0
    i dos punts d'aquesta recta són Q2 i Q3
    Els -3.875 haurien de ser 4.0, però amb el 4, el gràfic no surt be
    Pot ser és un bug de l'asy
*/
triple Q2 = (-3.875,-1,1.3125),
       Q3 = (-3.875,-3,2.9375);
dot(Q2,4.25bp+black);
draw(Q2--Q3,2bp+black,Arrow3(6));
label("$q_0$",Q2,-1.1X);
label("$\vec v$",(Q2+Q3)/2,-1.7X);
/*
    Dos punts de la perpendicular comuna
*/
triple R4 = (-1,0.615385,0),
       R5 = (2,0.615385,0);
dot(R4,4.25bp+red);      
draw(R4--R5,2bp+red,Arrow3(7));
label("$\vec u\times\vec v$",(R4+R5)/2,X+Z,red);    

/*
    Rectes R i R'
*/
label("$R$",P1,1.2Y,black);
label("$R'$",Q0,1.2X,black);
```

```python
%%asy --fmt pdf

settings.render = 10;
settings.prc = false;

import solids;
import graph3;

size(16cm,0);
currentprojection=orthographic(16,9,3.5);

real a = 1.35,
     t = 5.1,
     h = 0.0,
     m = 6.2,
     alpha = radians(1);
     
limits((0,0,0),(20,20,20));

real x(real t) { return 0.0; }
real y(real t) { return t; }
real z(real t) { return a * t; }
h = z(t);

draw((-12,0,0)--(12,0,0),red+linewidth(0.3mm),Arrow3(5));
draw((0,-12,0)--(0,12,0),red+linewidth(0.3mm),Arrow3(5));
draw((0,0,-12)--(0,0,12),red+linewidth(0.3mm),Arrow3(5));

draw(graph(x,y,z,-m,m,operator ..),black+linewidth(0.3mm));

path3 g = (0,t,h)..(t,0,h)..(0,-t,h)..(-t,0,h)..cycle;
draw(g,blue+linetype("4 4 4 4")+linewidth(0.3mm));

triple O = (0,0,h),
       P = (0,t,h),
       Q = (t * cos(alpha),t * sin(alpha),h);

dot(O,5bp+blue);
dot(P,5bp+black);
dot(Q,5bp+blue);

draw(P--O--Q,blue+linetype("0 4 4 4")+linewidth(0.3mm));

label("$p_0=(0,0,s)$",O,Z);
label("$p=(0,t,s)$",P,X+2.5Y-0.8Z);
label("$q=(x,y,z)$",Q,-X+Y-1.4Z);

revolution con = revolution(graph(new triple(real t) {
     return (0,t,a * t);},
      -m,m,25,operator ..),axis=Z);
draw(surface(con),opacity(0.3)+lightgreen);

label("$z$",(0,0,12.5),red);
label("$x$",(12.5,0,0),red);
label("$y$",(0,12.5,0),red);
```

```python
%%asy --fmt svg

import coordenades;

unitsize(1cm);

Canonica(-14,12,-8,14);

pen lx = linetype("8 8 8 8")+linewidth(0.1mm)+gray(0.5),
    ly = linewidth(0.4mm)+black;
pair p = (5,8),
     px = (p.x,0),
     py = (0,p.y);

draw(px--p--py,lx);
draw((0,0)--p,ly,Arrow(8));
draw((0,0)--(-2,3),ly,Arrow(8));

picture r = Coordenades(-5,5,-9,9,color=red,base=true,quadricula=false,ticky=0.1);
transform tr = (-2,3,1,-1,2,1);

pair pp = (4,-3),
    ppx = (pp.x,0),
    ppy = (0,pp.y);
lx =  linetype("8 8 8 8")+linewidth(0.1mm)+red;
ly = linewidth(0.4mm)+gray(0.6);
draw(r,ppx--pp--ppy,lx);
draw(r,(0,0)--pp,ly,Arrow(8));

add(tr * r);

dot("$p$",p,6bp+black);
```
